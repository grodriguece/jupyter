This dataset contains every weekly Hot 100 singles chart from Billboard.com. Each row of data represents a song and the corresponding position on that week's chart. Included in each row are the following elements:

- Billboard Chart URL
- WeekID
- Song name
- Performer name
- SongID - Concatenation of song & performer
- Current week on chart
- Instance (this is used to separate breaks on the chart for a given song. Example, an instance of 6 tells you that this is the sixth time this song has appeared on the chart)
- Previous week position
- Peak Position (as of the corresponding week)
- Weeks on Chart (as of the corresponding week)


```python
import pandas as pd
from plotnine import *
theme_set(theme_bw())

hot100_data = pd.read_csv('hot_stuff_2.csv', parse_dates=['weekid'], index_col=['weekid'])
hot100_data.head()
```

```python
hot100_data.dtypes
```

# One year of data

```python
hot100_data_2010 = hot100_data['2010']
hot100_data_2010.head()
```

```python
def highest_rank(x):
    return x.loc[x['week_position'].idxmin()][['weekid', 'week_position']]

hot100_data_2010 = hot100_data['2015']
hot100_data_2010 = hot100_data_2010[hot100_data_2010['week_position'] < 15]

number_weeks_in_data = hot100_data_2010[['song', 'performer', 'songid']].groupby(['song']).count()
song_often_occuring = number_weeks_in_data.where(number_weeks_in_data > 4).dropna().reset_index()['song']

plot_data = hot100_data_2010[hot100_data_2010['song'].isin(song_often_occuring)].reset_index()
highest_rank = plot_data.groupby(['song'], as_index=True).apply(highest_rank).reset_index()
```

```python
(
    ggplot(plot_data, aes(x='weekid', y='week_position')) + 
        geom_hline(aes(yintercept='yintercept'), pd.DataFrame({'yintercept': range(1,15)})) + 
        geom_line(aes(color='song')) + 
        geom_point(aes(color='song')) + 
        geom_text(aes(label='song'), data=highest_rank, size=8) + 
        scale_y_continuous(trans='reverse') + 
        scale_color_discrete(guide=False) +
        theme(figure_size=(22,6)) 
)
```

```python
adjusttext_settings = {'ha': 'right', 'expand_text': (1,1), 'arrowprops': {'arrowstyle': '->', 'color': 'red'}}
(
    ggplot(plot_data, aes(x='weekid', y='week_position')) + 
        geom_hline(aes(yintercept='yintercept'), pd.DataFrame({'yintercept': range(1,15)})) + 
        geom_line(aes(color='song')) + 
        geom_point(aes(color='song')) + 
        geom_text(aes(label='song'), data=highest_rank, size=8, adjust_text=adjusttext_settings) + 
        scale_y_continuous(trans='reverse') + 
        scale_color_discrete(guide=False) +
        theme(figure_size=(22,6)) 
)
```

# Multiple years of data

```python
def highest_rank(x):
    return x.loc[x['week_position'].idxmin()][['weekid', 'week_position']]

hot100_data_subset = hot100_data['2014':'2015']
hot100_data_subset = hot100_data_subset[hot100_data_subset['week_position'] < 15]

number_weeks_in_data = hot100_data_subset[['song', 'performer', 'songid']].groupby(['song']).count()
song_often_occuring = number_weeks_in_data.where(number_weeks_in_data > 4).dropna().reset_index()['song']

plot_data = hot100_data_subset[hot100_data_subset['song'].isin(song_often_occuring)].reset_index()
#plot_data['Song'] = plot_data['Song'].str.replace('$', 's')
plot_data['year'] = plot_data['weekid'].dt.year
highest_rank = plot_data.reset_index().groupby(['song', 'year'], as_index=True).apply(highest_rank).reset_index()

adjusttext_settings = {'ha': 'right', 'expand_text': (1,1), 'arrowprops': {'arrowstyle': '->', 'color': 'red'}}
(
    ggplot(plot_data, aes(x='weekid', y='week_position')) + 
        geom_hline(aes(yintercept='yintercept'), pd.DataFrame({'yintercept': range(1,15)})) + 
        geom_line(aes(color='song')) + 
        geom_point(aes(color='song')) + 
        #geom_text(aes(label='Song'), size=8) + 
        geom_text(aes(label='song'), data=highest_rank, size=8, adjust_text=adjusttext_settings) + 
        scale_y_continuous(trans='reverse') + 
        scale_color_discrete(guide=False) + 
        facet_wrap('~ year', ncol=1, scales='free') +
        theme(figure_size=(22,13)) 
)
```
