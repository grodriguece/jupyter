# fixing bar size with ggplot in facet_grid()
[link](https://stackoverflow.com/questions/59390543/fixing-bar-size-with-ggplot-in-facet-grid/63218246#63218246)


```python
import numpy as np
from plotnine import *
import pandas as pd
```

```python
x = list(range(1, 6))
df = pd.DataFrame({'A' : [chr(x) for x in range(65, 65+17)], 'B' : np.random.choice(500, 17).tolist(), 
                   'C' : np.repeat(x, [5, 6, 2, 2, 2]).tolist()})
df
```

```python
(
    ggplot(data=df, mapping = aes(x = 'C',  y = 'B', label = 'A', fill = 'A')) + 
    geom_bar(stat = "identity", position = "dodge") + 
    geom_text(position = position_dodge(width = 0.9), angle = 90)
)
```

```python
(
    ggplot(data=df, mapping = aes(x = 'C',  y = 'B', label = 'A', fill = 'A')) + 
    geom_bar(stat = "identity", position = "dodge") + 
    geom_text(position = position_dodge(width = 0.9), angle = 90) +
    facet_grid('~C')
)
```

```python
(
    ggplot(data=df, mapping = aes(x = 'C',  y = 'B', label = 'A', fill = 'A')) + 
    geom_col(position = position_dodge2(width = 0.9, preserve = 'single')) + 
    geom_text(position = position_dodge(width = 0.9), angle = 90)
)

# ggplot(data,aes(x = C,  y = B, label = A, fill = A)) +
#   geom_col(position = position_dodge2(width = 0.9, preserve = "single")) +
#   geom_text(position = position_dodge2(width = 0.9, preserve = "single"), angle = 90, vjust=0.25)
```

# As per plotnine documentation
 ![Tux, the Linux mascot](tux.png)

spacestr in ['fixed', 'free', 'free_x', 'free_y'] Whether the x or y sides of the panels should have the size. It also depends to the scales parameter. Default is 'fixed'. This setting is not yet supported.
[plotnine.facets.facet_grid](https://plotnine.readthedocs.io/en/stable/generated/plotnine.facets.facet_grid.html?highlight=facet_grid#plotnine.facets.facet_grid)

```python
(
    ggplot(data=df, mapping = aes(x = 'A',  y = 'B', fill = 'A')) + 
    geom_col(position = 'dodge') +
    facet_grid('~C', scales = "free_x", space = "free_x")
)
```
